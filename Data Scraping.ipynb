```python
### Function to scrap 1st page
```

```python
import pandas as pd
def quotes(soup):
    author_name = [i.text.strip()for i in soup.find_all('span',class_ = 'authorOrTitle')]
    quotes = [list (i)[0].strip() for i in soup.find_all('div',class_ = 'quoteText')]
    likes = [int(i.text.strip().split()[0]) for i in soup.find_all('a',class_='smallText')]
    return pd.DataFrame({'Authors':author_name,'Quotes':quotes,'Likes':likes})
```

```python
df = quotes(soup)
df
```

```python
df.to_csv('quotes.csv')
```

```python
stud = 'https://www.goodreads.com/quotes?page={}'
```

```python
tables = []
from tqdm import tqdm
for i in range(1,51):
    url_all = stud.format(i)
    html_all = requests.get(url_all).text
    soup = BeautifulSoup(html_all,'html.parser')
    tables.append(quotes(soup))         
```

```python
data =pd.concat(tables,axis = 0)
data.head()
```
